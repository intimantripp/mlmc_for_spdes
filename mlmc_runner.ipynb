```python
import numpy as np
from scipy.stats import norm
```

```python
def call_bsm(S0, K, sigma, T, r):

    d1 = (np.log(S0/K) + (r + sigma**2/2)*T)/(sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)

    c = S0 * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)

    return c

call_bsm(100, 100, 0.2, 1, 0.05)
```
