```python
import numpy as np
from mlmc.stoch_heat_eqn_qoi import stoch_heat_eqn_qoi

## Calculate nth Fourier mode of the solution u ##

def nth_fourier_mode(n, u):
    # Calculates integral from 0 to 1 of 2*u(x,t)sin(n pi x) dx
    x = np.linspace(0, 1, u.shape[0])
    sin_basis = np.sin(n * np.pi * x)[:, np.newaxis]
    integrand = 2 * u * sin_basis
    fourier_mode = np.trapz(integrand, x, axis=0)
    return fourier_mode

def suspect_noise_coupling(dWf):
    num_rows = int(dWf.shape[0] / 2 - 0.5) # nf - 1 internal points for fine. Corresponds to (nf + 1) / 2 - 1 coarse internals
    dWc =  dWf[:-1, :].reshape(num_rows, 2, dWf.shape[1]).sum(axis=1) * 0.5 # sum of 2 fine adjacent increments, ignore final point
    return dWc

first_fourier = lambda u: nth_fourier_mode(1, u)
stoch_heat_eqn_qoi(qoi_fn=first_fourier, validation_value=0.0, noise_coupling=suspect_noise_coupling)
```

```python
import numpy as np
from mlmc.stoch_heat_eqn_qoi import stoch_heat_eqn_qoi
def nth_fourier_mode(n, u):
    # Calculates integral from 0 to 1 of 2*u(x,t)sin(n pi x) dx
    x = np.linspace(0, 1, u.shape[0])
    sin_basis = np.sin(n * np.pi * x)[:, np.newaxis]
    integrand = 2 * u * sin_basis
    fourier_mode = np.trapz(integrand, x, axis=0)
    return fourier_mode

first_fourier_var = lambda u: nth_fourier_mode(1, u)**2
true_var = lambda n: (1 - np.exp(- 2 * n**2 * np.pi**2 * 0.25)) / (n**2 * np.pi**2)

stoch_heat_eqn_qoi(qoi_fn=first_fourier_var, validation_value=true_var(1))
```
